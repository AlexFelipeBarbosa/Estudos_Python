```python
# Importando a Biblioteca - Pandas
import pandas as pd
import numpy as np
```

```python
# Carregar dados de um arquivo csv
df = pd.read_csv('dados/df_vendas.csv')
df_lojas = pd.read_csv('dados/df_Lojas.csv')
df_produtos = pd.read_csv('dados/df_Produtos.csv')
```

```python
# Mudar tipo da coluna
df_lojas["Numero da loja"] = df_lojas["Numero da loja"].astype(str)
df_produtos["Numero do produto"] = df_produtos["Numero do produto"].astype(str)
df["Numero da loja"] = df['Numero da loja'].astype(str)
df["Numero do produto"] = df["Numero do produto"].astype(str)
```

```python
# Mesclando usando left join
df_vendas = df.merge(df_lojas, on="Numero da loja", how="left")
df_vendas = df_vendas.merge(df_produtos, on="Numero do produto", how='left')
df_vendas
```

```python
# Deletando colunas desnecessarias 
df_vendas = df_vendas.drop(["Numero da loja", "Numero do produto"],axis=1)
df_vendas
```

```python
# Reordenando colunas 
df_vendas.columns
df_vendas = df_vendas[['Nome da loja','Cidade', 'Estado', 'Canal de vendas', 'Nome do produto',
                      'Categoria do produto', 'Marca do produto', 'Preco do produto',
                      'Mes da venda', 'Ano da venda', 'Quantidade vendida', ]]
df_vendas
```

```python
# Criar nova colunas a partir de outros valores 
df_vendas["Preco final"] = df_vendas["Quantidade vendida"] * df_vendas["Preco do produto"]
df_vendas
```

```python
df_vendas
```

```python
# Agrupando por colunas (e arredondar valor)
df_vendas.groupby(["Estado","Cidade"])["Preco final"].sum()
```

```python
# Agrupando por colunas (e arredondar valor)
df_vendas.groupby("Categoria do produto")[["Preco final","Quantidade vendida"]].sum()
```

```python
# Tabela dinamica 
df_pivot = df_vendas.pivot_table(values="Preco final", index=["Canal de vendas", "Nome da loja"], columns="Mes da venda", aggfunc="sum")
df_pivot
```
