```python
# Renomear colunas 
df.columns
df.rename(columns={"Quantidade vendida":"QTD", "Ano da venda":"Ano"}, inplace=True)
df.head()
```

```python
# Operacoes com as colunas 
df["QTD"].max() # Valor maximo
df["QTD"].min() # Valor minimo
df["QTD"].sum() # Valor somado
df["QTD"].mean() # media
df["QTD"].count() # contagem
df["QTD"].unique() # valores unicos
df["QTD"].value_counts() # quantas vezes cada valor foi vendido
```

```python
# Ordenar df com base em uma coluna
df.sort_values(["QTD","Mes da venda"], ascending=False) # quantidade decrescente e por mes
```

```python
# filtrar coluna com base em valor e condicionais
df[df["QTD"]>= 40] # filtrar as quantidades >= 40
```

```python
df[df["QTD"]== 20] # vendas que a quantidade é igual a 20
```

```python
# Filtrando com duas condicoes "and"
df[(df["QTD"]== 20) & (df["Mes da venda"] == 1)] # Quatidade = 20 somente do mes 1
```

```python
# Filtrando com duas condicoes "or"
df[(df["QTD"]== 20) | (df["Mes da venda"] == 2)].sample(7) # sample = linhas aleatorias 
```

```python
# Verificar valores nulos (em branco)
df.isnull().sum()
```

```python
# Filtrar linhas em branco
df[df["QTD"].isnull()]
```

```python
# Deletar linhas em branco
df_sem_nulos = df.dropna()
df_sem_nulos.info()
```

```python
# Preencher valores em branco com outros valores 
df["QTD"].mean() # media
df["QTD"].median() # mediana (valor que mais aparece)

df = df.fillna(df["QTD"].median())
df.info()
```

```python
# Agrupar valores selecionados
df1 = df.groupby(["Ano","Mes da venda"], as_index=False).aggregate({"QTD":["mean", "sum"]})
df1
```

```python
# Tabela dinamica
df.pivot_table(values="QTD", index="Numero da loja", columns="Mes da venda", aggfunc="sum")
```

```python
# Tabela dinamica
df_pivot = df.pivot_table(values="QTD", index="Numero da loja", columns=["Ano","Mes da venda"], aggfunc="sum")
df_pivot
```
